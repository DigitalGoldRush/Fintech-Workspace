```python
# Import the required libraries and dependencies
import pandas as pd
from pathlib import Path
import hvplot.pandas
```

```python
# Using the Pandas read_csv function and the Path module, read in the data
# from the "state_loan_data.csv" file and create the Pandas DataFrame
# Using the Pandas read_csv function and the Path module, read in the data
# from the "state_loan_data.csv" file and create the Pandas DataFrame
loan_data_df = pd.read_csv(
    Path("/Users/michaeldionne/Documents/GitHub/Fintech-Workspace/Module 6/Class_2_Activities/07-Stu_Picture_Perfect/Resources/state_loan_data.csv")
)

# Remove all null values from the loan data
loan_data_df = loan_data_df.dropna()

# Review the first and last five rows of the DataFrame
display(loan_data_df.head())
display(loan_data_df.tail())
```

```python
# Group and sum the data by "State Code" for state-level analysis
state_loan_data_df = loan_data_df.groupby('State Code').sum()

# Review the first and last five rows of state-level data
display(state_loan_data_df.head())
display(state_loan_data_df.tail())
```

```python
# Create a Series by accessing the data from the "Total Average Loan Amount" column
# Sort the values in the `total_average_loan_amount_series`
total_average_loan_amount = state_loan_data_df['Total Average Loan Amount']
total_average_loan_amount_sorted = total_average_loan_amount.sort_values()

# Create the plot for the `total_average_loan_amount_sorted`
plot_total_average_loan_amount_sorted = total_average_loan_amount_sorted.hvplot.bar(
    title ='Total Average Loan Amount by State',
    color="blue",
    hover_color = "yellow",
    xlabel='State Code',
    ylabel="Total Average Loan Amount",
    rot=90
).opts(yformatter='%.0f')

# Visualize the plot
plot_total_average_loan_amount_sorted
```

```python
# Create a Series by accessing the data from the "2015-2016" column
# Sort the values in the `average_loan_2015_2016`
average_loan_2015_2016 = state_loan_data_df['2015 - 2016']
average_loan_2015_2016_sorted = average_loan_2015_2016.sort_values()

# Create the plot for the `average_loan_2015_2016_sorted`
plot_average_loan_2015_2016_sorted = average_loan_2015_2016_sorted.hvplot.bar(
    title='Average Loan Amount - 2015 - 2016',
    color="green",
    hover_color="yellow",
    xlabel='State Code',
    ylabel="Average Loan Amount",
    rot=90
).opts(yformatter='%.0f')

# Visualize the plot
plot_average_loan_2015_2016_sorted
```

```python
# Create a Series by accessing the data from the "2010-2014" column
# Sort the values in the `average_loan_2015_2016`
average_loan_2010_2014 = state_loan_data_df['2010 - 2014']
average_loan_2010_2014_sorted = average_loan_2010_2014.sort_values()

# Create the plot for the `average_loan_2010_2014_sorted`
plot_average_loan_2010_2014_sorted = average_loan_2010_2014_sorted.hvplot.bar(
    title='Average Loan Amount - 2010 - 2014',
    color="orange",
    hover_color="yellow",
    xlabel='State Code',
    ylabel="Average Loan Amount",
    rot=90
).opts(yformatter='%.0f')

# Visualize the plot
plot_average_loan_2010_2014_sorted
```

```python
# Compose a side-by-side plot using the '+' operator
# and the variable names for the 2010-2014 and 2015-2016 plots
plot_average_loan_2010_2014_sorted + plot_average_loan_2015_2016_sorted
```

```python
# A composition placing 3 plots side-by-side using the '+' operator
# and the variable names for the total, 2010-2014, and 2015-2016 plots
plot_total_average_loan_amount_sorted + plot_average_loan_2010_2014_sorted + plot_average_loan_2015_2016_sorted

```

```python
# Compose an overlay plot using the '*' operator
overlay_plot = plot_average_loan_2010_2014_sorted * plot_average_loan_2015_2016_sorted

# Style the new overlay plot
overlay_plot.opts(
    height= 500,
    width=1000,
    xlabel="State Code",
    ylabel="Average Loan Amount",
    title ="Comparison of Average Loan Amounts - 2010-2014 vs. 2015-2016"
)
```
